```python
#importing the tkinter module
import tkinter as tk

def calculate(): #defines a function called calculate(). This function takes in two integers and an operator as parameters.

    first_number = int(first_number_entry.get())

    second_number = int(second_number_entry.get())

    operator = operator_entry.get()
#The first number is stored in the first_number variable and the second number is stored in second_number.
#The code sets up a conditional statement to check the operator of the input. 
#The first number is stored in the first_number variable and the second number is stored in second_number. 
#if operator == "1": result = first_number + second_number result_label.config(text="Result:" + str(result))

    if operator == "1":

        result = first_number + second_number

        result_label.config(text="Result:" + str(result))

    elif operator == "2":

        result = first_number - second_number

        result_label.config(text="Result:" + str(result))

    elif operator == "3":

        result = first_number * second_number

        result_label.config(text="Result:" + str(result))

    elif operator == "4":

        result = first_number / second_number

        result_label.config(text="Result:" + str(result))
        
    elif operator == "5":
        
        result = first_number ** second_number

        result_label.config(text="Result:" + str(result))
        
    else:

        result_label.config(text="Invalid Operator")
        
#The operator parameter stores what operation to perform on these numbers: "1" for addition, "2" for subtraction, "3" for multiplication, or "4" for division or "5 for power". 
#If the operator parameter equals 1 then result will be calculated using addition; if it equals 2 then result will be calculated using subtraction; if it equals 3 then result will be calculated using multiplication; and if it equals 4 then result will be calculated using division.

#GUI using TKinter

root = tk.Tk()

root.title("Calculator")

#root.geometry("820x1024")

 

first_number_label = tk.Label(root, text="First number")

first_number_label.pack()

first_number_entry = tk.Entry(root)

first_number_entry.pack()
# the lines create a label and entry widget for the first number.
#The label is set to display the text 'first number', and the entry widget is used for entering the first number.
#both the label and the entry widget are created as children of the root window and arranged using the pack method.
 

second_number_label = tk.Label(root, text="Second number")

second_number_label.pack()

second_number_entry = tk.Entry(root)

second_number_entry.pack()


operator_label = tk.Label(root, text="Operator (1-Addition, 2-Subtraction, 3-Multiplication, 4-Division, 5-Power)")

operator_label.pack()

operator_entry = tk.Entry(root)

operator_entry.pack()

 

calculate_button = tk.Button(root, text="Calculate", command=calculate)

calculate_button.pack()

 

result_label = tk.Label(root, text="Result")

result_label.pack()

 

root.mainloop()
#The code creates a window that is 820x1024 pixels in size. 
#The label widget has the text "First number" and the entry widget has an empty string as its text.
 
```
