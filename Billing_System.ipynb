```python
items = {
    10 : ['Apple','Bannana','Orange','Pear'],
    20 : ['Dragon Fruit','Milk','Curd','Lassi','Thumbs Up','Pepsi','Cocacola','Fanta','Marinda'],
    35 : ['Glassiy','Vannila Cream'],
}
```

```python
def billing_system(items):
    total_cost = 0
    bill_items = []  # To store details of each item for the bill
    serial_number = 1  # Initialize serial number
    total_quantity = 0  # Initialize total quantity
    
    print("Welcome to the Billing System")
    print("=============================")
    
    while True:
        user_input = input('Enter item (type "done" to finish): ').strip().capitalize()
        
        if user_input == 'Done':
            break
        
        quantity = int(input('Enter quantity: '))
        
        found = False
        for price, products in items.items():
            if user_input in products:
                item_total = price * quantity
                total_cost += item_total
                bill_items.append((serial_number, user_input, quantity, price, item_total))
                serial_number += 1
                total_quantity += quantity
                found = True
                break
        
        if not found:
            print('Item not found in the list.')
    
    # Print the bill
    print("\n=========== Your Bill ===========")
    for item in bill_items:
        print(f"{item[0]}. {item[1]} x {item[2]}: ${item[4]}")
    
    print("---------------------------------")
    print(f"Total Quantity: {total_quantity}")
    print(f"Total: ${total_cost}")
    print("=================================")
    print('***Thank You for Shopping***')
    print('***Pls Visit Again***')

    return 0
```

```python
import tkinter as tk
from tkinter import messagebox

# Define the billing system function
def billing_system(items):
    total_cost = 0
    bill_items = []
    serial_number = 1
    total_quantity = 0
    
    # Function to handle 'Add Item' button click
    def add_item():
        nonlocal total_cost, bill_items, serial_number, total_quantity
        user_input = item_entry.get().strip().capitalize()
        quantity = quantity_entry.get().strip()
        
        if not quantity.isdigit():
            messagebox.showerror("Error", "Quantity must be a number.")
            return
        
        quantity = int(quantity)
        
        found = False
        for price, products in items.items():
            if user_input in products:
                item_total = price * quantity
                total_cost += item_total
                bill_items.append((serial_number, user_input, quantity, price, item_total))
                serial_number += 1
                total_quantity += quantity
                found = True
                break
        
        if not found:
            messagebox.showerror("Error", "Item not found in the list.")
        
        # Clear entry fields after adding item
        item_entry.delete(0, tk.END)
        quantity_entry.delete(0, tk.END)
    
    # Function to handle 'Finish' button click
    def finish():
        # Print the bill
        bill_output.delete('1.0', tk.END)  # Clear previous bill
        bill_output.insert(tk.END, "=========== Your Bill ===========\n")
        for item in bill_items:
            bill_output.insert(tk.END, f"{item[0]}. {item[1]} x {item[2]}: ${item[4]}\n")
        bill_output.insert(tk.END, "---------------------------------\n")
        bill_output.insert(tk.END, f"Total Quantity: {total_quantity}\n")
        bill_output.insert(tk.END, f"Total: ${total_cost}\n")
        bill_output.insert(tk.END, "=================================\n")
        bill_output.insert(tk.END, "***Thank You for Shopping***\n")
        bill_output.insert(tk.END, "***Pls Visit Again***")
        
        # Disable 'Add Item' and 'Finish' buttons after finishing
        add_button.config(state=tk.DISABLED)
        finish_button.config(state=tk.DISABLED)
    
    # Create GUI window
    root = tk.Tk()
    root.title("Billing System")
    
    # GUI elements
    tk.Label(root, text="Welcome to the Billing System").pack()
    tk.Label(root, text="=============================").pack()
    
    item_label = tk.Label(root, text="Enter item: ")
    item_label.pack()
    item_entry = tk.Entry(root, width=30)
    item_entry.pack()
    
    quantity_label = tk.Label(root, text="Enter quantity: ")
    quantity_label.pack()
    quantity_entry = tk.Entry(root, width=30)
    quantity_entry.pack()
    
    add_button = tk.Button(root, text="Add Item", command=add_item)
    add_button.pack(pady=10)
    
    finish_button = tk.Button(root, text="Finish", command=finish)
    finish_button.pack(pady=10)
    
    bill_output = tk.Text(root, height=15, width=50)
    bill_output.pack(pady=20)
    
    root.mainloop()

# Run the billing system GUI
billing_system(items)
```
